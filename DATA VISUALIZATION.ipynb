```python

import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
```

```python

year = [2014,2015,2016,2017,2018]
sales = [1234,4344,4454,3222,3343]
profit = [123,443,445,432,222]
```

```python
plt.plot(year, sales, 'red')
plt.xlabel('Year')
plt.ylabel('Sales')
plt.title('Sales vs Year')
```
